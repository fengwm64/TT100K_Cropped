```python
import json
import os
import random
import numpy as np
import cv2
import matplotlib.pyplot as plt
import anno_func

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

# 1. 读取TT100K数据集信息

```python
# 定义数据目录和文件路径
datadir = "G:\data"
filedir = os.path.join(datadir, "annotations.json")

# 读取并解析注释文件
with open(filedir, 'r') as f:
    annos = json.load(f)

# 'ids.txt' 包含训练图像的标识列表
train_ids = open(os.path.join(datadir, "train/ids.txt")).read().splitlines()
test_ids = open(os.path.join(datadir, "test/ids.txt")).read().splitlines()
```

# 2. 每个类别提取固定数量图片

```python
# 训练集
train_output_dir = "./cropped_images/train"
anno_func.crop_and_save_images(annos, datadir, train_ids, train_output_dir, num_per_class=300)

# 测试集
test_output_dir = "./cropped_images/test"
anno_func.crop_and_save_images(annos, datadir, test_ids, test_output_dir, num_per_class=300)
```

# 3. 提取每个类别所有图片

```python
# 训练集
train_output_dir = "./cropped_images/train"
train_class_count = anno_func.crop_and_save_images(annos, datadir, train_ids, train_output_dir)

# 测试集
test_output_dir = "./cropped_images/test"
test_class_count = anno_func.crop_and_save_images(annos, datadir, test_ids, test_output_dir)
```

```python
# 绘制统计图
anno_func.plot_class_distribution(train_class_count)
anno_func.plot_class_distribution(test_class_count)
```
